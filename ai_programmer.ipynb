````python
import erniebot as eb  
from ipywidgets import widgets, Layout, Text, Button, Output, HBox, VBox, Textarea 
from IPython.display import clear_output,HTML 
import os  
import re 
import subprocess


def ai_programmer(history=[]):
    global_namespace = {}
    
    # 创建界面元素  
    role_set = Textarea(value='你是一个专业的python程序员，请按以下要求开发python程序：', layout={'width': '99%'})  
    input_text = Textarea(placeholder='在此输入您的编程需求', layout={'width': '99%','height': '100px'})  
    out_code = Textarea(placeholder='本次生成的代码', layout={'width': '99%','height': '100px'}, disabled=True) 
    submit_button = Button(description='提交', button_style='success')  
    clear_button = Button(description='清除', button_style='info') 
    output = Output(layout={'border': '1px solid black','width': '99%'},scroll_output=True)  

    code_type = widgets.Dropdown(
            options=["python", "java"],
            value="python",
            layout={'width': '190px'},
            description="代码类型"
        )

    operate_mode = widgets.Dropdown(
            options=["只有出错时将错误信息自动填写输入框", "每次执行结果自动填写在输入框",  "不进行自动填入操作"],
            value="只有出错时将错误信息自动填写输入框",
            layout={'width': '190px'},
            description="自动填写模式"
        )

    clear_mode = widgets.Dropdown(
            options=["点击按钮清理输出框","每次提交都清理输出框"],
            value="点击按钮清理输出框",
            layout={'width': '190px'},
            description="清理模式"
        )

    history_mode = widgets.Dropdown(
            options=["每次交互包括历史记录","不上传历史记录"],
            value="每次交互包括历史记录",
            layout={'width': '190px'},
            description="历史模式"
        )

    def run_java_code(java_code):
        # 将Java代码保存到一个临时文件
        java_filename = "TemporaryJavaCode.java"
        class_filename = "TemporaryJavaCode.class"
        jdk_path = "/home/aistudio/work/jdk-21.0.1/bin/"
        try:
            with open(java_filename, "w") as java_file:
                java_file.write(java_code)

            # 尝试编译Java代码
            subprocess.run([jdk_path+'javac', java_filename], check=True, stderr=subprocess.PIPE)

        except subprocess.CalledProcessError as e:
            # 获取编译错误信息，并将其raise出去
            error_message = e.stderr.decode()
            raise RuntimeError(f"Java code compilation failed. Error message:\n{error_message}")
            return 1
        finally:
            # 清理临时文件
            os.remove(java_filename)

        try:
            # 尝试执行Java程序
            execute_process = subprocess.run([jdk_path+'java', java_filename[:-5]], check=True, stderr=subprocess.PIPE)
        except subprocess.CalledProcessError as e:
            # 获取编译错误信息，并将其raise出去
            error_message = e.stderr.decode()
            raise RuntimeError(f"Java code run failed. Error message:\n{error_message}")
            return 2
        finally:
            # 清理临时文件
            os.remove("TemporaryJavaCode.class")
        
        return 0


    # 定义 API 调用函数  
    def get_yiyan(question, token): 
        eb.api_type = 'aistudio'  
        eb.access_token = token  
        history.append({'role': 'user', 'content': question})
        response = eb.ChatCompletion.create(  
            model='ernie-bot',  
            messages=history,  
        )
        history.append({'role': 'assistant', 'content': response.result})  
        return response.result  

    def extract_code(text):  
        # 使用正则表达式匹配代码块
        if code_type.value=='java':
            pattern = r'```java.*?\n(.*?)```'  
        else:
            # 默认python
            pattern = r'```python.*?\n(.*?)```'  
        matches = re.findall(pattern, text, re.DOTALL)  
        if matches:  
            # 如果匹配成功，返回第一个匹配到的代码块  
            return matches[0]  
        else:  
            # 如果匹配失败，返回空字符串  
            return "" 

    # 定义提交按钮的回调函数  
    def on_submit(b):  
        global history
        if history_mode=='不上传历史记录':
            history=[]
        last_result = ""  
        '''
        if history==[]:
            question = role_set.value + input_text.value 
        else:
            question = input_text.value
        '''
        question = role_set.value + input_text.value 

        with output: 
            if clear_mode.value=='每次提交都清理输出框': 
                clear_output()  
            print("-------------------------------问题内容-------------------------------\n"+question) 
            print("处理中，请稍等......")
            response = get_yiyan(question, my_token)  # 用您的 token 替换 'YOUR_TOKEN'   
            print("-------------------------------返回内容-------------------------------\n"+response)
            the_code = extract_code(response)
            out_code.value=the_code
            #print("抽取代码：\n"+the_code)
            if the_code!="":
                try:
                    print("-------------------------------执行结果-------------------------------\n")
                    if code_type.value=='java': 
                        run_java_code(the_code)   
                    else:
                        exec(the_code,global_namespace)
                    print("\n----------------------------------------------------------------------\n")  
                    last_result="代码执行成功！"
                    if operate_mode.value=='每次执行结果自动填写在输入框':
                        input_text.value=last_result
                except Exception as e:  
                    last_result="代码执行出错，错误信息：\n"+str(e)
                    if operate_mode.value!='不进行自动填入操作':
                        input_text.value=last_result
            else:
                last_result="没有生成可执行代码"
                if operate_mode.value!='不进行自动填入操作':
                    input_text.value=last_result
            print(last_result)

    # 定义发送错误信息的回调函数  
    def on_clear_send(b):  
        input_text.value=''
        history=[]
        out_code.value=''
        with output:  
            clear_output()  

    def on_code_change(b):
        global history
        history=[]
        #使用的时候“Java的执行类名为TemporaryJavaCode”这句必须包括
        if code_type.value=='java':
            role_set.value = "你是一个专业的java程序员，请按以下要求开发java程序，首先Java的执行类名为TemporaryJavaCode，其他的要求包括："
        else:
            # 默认python
            role_set.value = "你是一个专业的python程序员，请按以下要求开发python程序："
        out_code.value=''
        with output:  
            clear_output()  


    # 绑定事件触发函数  
    submit_button.on_click(on_submit)  
    clear_button.on_click(on_clear_send)
    code_type.observe(on_code_change)

    # 整合界面并显示  
    title = widgets.HTML(value="<h1 style='text-align: center;'>编程小助手（Python）</h1>")  
    main_layout = VBox([
        title,
        role_set,
        input_text,   
        HBox([code_type,operate_mode,clear_mode,history_mode,submit_button, clear_button]), 
        out_code,
        output])
    # 显示界面
    display(main_layout)   
````
